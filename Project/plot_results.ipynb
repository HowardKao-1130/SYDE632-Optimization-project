```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the Excel file into a pandas DataFrame
# df = pd.read_csv('run_results_on_gender_dataset_with_seed_78_and_KKT_at_2024-04-08_18-38-35.csv')
df = pd.read_csv('run_results_on_cancer_dataset_with_seed_75_and_KKT_at_2024-04-08_18-42-10.csv')

# Extract the relevant columns
column1 = df.iloc[:, 0]
column3 = df.iloc[:, 3]
column4 = df.iloc[:, 4]
column5 = df.iloc[:, 5]
column6 = df.iloc[:, 6]

label3 = 'Majority class accuracy in train set'
label4 = 'Minority class accuracy in train set'
label5 = 'Majority class accuracy in test set'
label6 = 'Minority class accuracy in test set'

# Plot the columns against column 1
plt.figure(1)
plt.plot(column1, column3, label=label3)
plt.plot(column1, column4, label=label4)
plt.plot(column1, column5, label=label5)
plt.plot(column1, column6, label=label6)
plt.xlabel('Number of Updates')
plt.ylabel('Accuracy')
plt.title('Classification Accuracy versus Number of Updates')
plt.legend()

plt.figure(2)
plt.plot(column1, column3, label=label3)
plt.plot(column1, column4, label=label4)
plt.xlabel('Number of Updates')
plt.ylabel('Train Set Accuracy')
plt.title('Train Set Classification Accuracy versus Number of Updates')
plt.legend()

plt.figure(3)
plt.plot(column1, column5, label=label5)
plt.plot(column1, column6, label=label6)
plt.xlabel('Number of Updates')
plt.ylabel('Test Set Accuracy')
plt.title('Test Set Classification Accuracy versus Number of Updates')
plt.legend()

# Set labels and title

# Display the legend

# Show the plot
plt.show()
```
